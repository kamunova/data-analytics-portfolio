```python
# Задание 1- чтение файла
```

```python
lines=[]

try:
    with open("users.txt","r", encoding="utf-8") as file:
        lines=file.readlines()
except FileNotFoundError:
    print("the file that you would like to read is not found")
except:
    print("some error occured")
print(lines)
```

```python
# Задание 2- класс Users
```

```python
class Users:
    def __init__(self,name,surname,email, birth_year):
        self.name=name
        self.surname= surname
        self.email=email
        self.birth_year=int(birth_year)
    def age(self):
        return 2023 - self.birth_year
```

```python
u=Users("Magzhan","Kairanbay","magzhan.kairanbay@gmail.com",1991)
print(u.name, u.surname,u.email)
print(u.age())
```

```python
# Задание 3 - разделение данных
```

```python
users_list=[]
for line in lines:
    data=line.strip().split(",")
    if not data[4].isdigit():
        continue
    user= Users(data[1],data[2],data[3],data[4])
    users_list.append(user)
for user in users_list:
    print(user.name,user.surname,user.email,user.age())
```

```python
# Задание 4- фильтр
```

```python
def filter(users, low=30,high=50):
    for user in users:
        age=user.age()
        if low<age<high:
            print(user.email,age)
```

```python
filter(users_list)
```
